```python
#この内容をもとにカレントディレクトリを検索してみる
#１行目globを読み込む
#２行目検索キーに該当するファイルを全て取得し、tbl_order_files変数に格納している

import glob
tbl_order_files = glob.glob(tbl_order_file)
tbl_order_files
```

### ４　複数データを結合（ユニオン）

```python
order_all = pd.DataFrame()
#処理結果を格納するデータフレームを用意
file = tbl_order_files[0]
#リスト内の１個目のファイルを処理対象として指定
order_data = pd.read_csv(file)
#注文データファイルを読み込み
print(f'{file}:{len(order_data)}')
#読み込んだファイル名と件数を出力
#print(f'[file]:{len(order_data)}')は、表示フォーマットを指定する書き方
#{}の中にそもまま変数を記載できる。ファイルのフルパスと入力レコード数が：で区切って表示される
order_all = pd.concat([order_all,order_data],ignore_index=True)
#ユニオンする
order_all
```

### 繰り返し処理を実行

```python
order_all = pd.DataFrame()
for file in  tbl_order_files:
    order_data = pd.read_csv(file)
    print(f'{file}:{len(order_data)}')
    order_all = pd.concat([order_all,order_data],ignore_index=True)
order_all
#tbl_order_filesから１個づつ取り出してfileに格納し３〜５行目の処理を行う
#４行目でファイル名と件数を出す事でどのファイルから何件処理されたのかをみることができる
```

## ノック６　不要なデータを除外

```python
# store_id:９９９の注文データを除外する場合の書き方
order_data = order_all.loc[order_all['store_id']!=999]
#999以外のデータを抽出して、order_data変数に格納
order_data
```

### ７　マスタデータを結合　

```python
order_data = pd.merge(order_data,m_store,on='store_id',how='left')
#order_dataとm_store（店舗マスタ情報：店舗名など）をジョインする
#この時共通する列はstore_idになる。共通する列をon=に指定することで紐付き値を横に結合することができる
#how=でoder_dataとm_soreのどちらを主軸にするかを指定
#今回はorder_dataに店舗マスタの情報を連結することにしたのでhow='left'と記述
order_data
```

## ８マスタが存在しないコードに名称を設定

```python
order_data = pd.merge(order_data, m_area, on='area_cd', how='left')
order_data
#area_cdは、店舗マスタをジョインして追加される項目なので店舗マスタをジョインするまでエリアマスタはジョインでできない。順番とキーを事前に確認しておく
```

```python
#takeout_flagとstatusの記号が何を意味しているのかwからないので、記号に名称を設定する
#まずは、order_dataにtakeout_flag項目を追加してそれぞれに名称を設定する
```

```python
order_data.loc[order_data['takeout_flag'] == 0, 'takeout_name'] ='デリバリー'
order_data.loc[order_data['takeout_flag'] == 1, 'takeout_name'] ='お持ち帰り'
order_data
```

```python
order_data.loc[order_data['status'] == 0, 'status_name'] ='受付'
order_data.loc[order_data['status'] == 1, 'status_name'] ='お支払い済み'
order_data.loc[order_data['status'] == 2, 'status_name'] ='お渡し済み'
order_data.loc[order_data['status'] == 9, 'status_name'] ='キャンセル'
order_data
```
